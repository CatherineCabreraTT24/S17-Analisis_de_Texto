# Guía Práctica: Preprocesamiento y Vectorización de Textos (NLP)
Este notebook contiene la implementación directa y limpia de los conceptos fundamentales vistos en la presentación.

## Paso 0: Instalación y Carga de Librerías

```python
#nlp --> Natural language processing

import re
import nltk
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer, PorterStemmer
from nltk.corpus import stopwords
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer

# Descargar recursos esenciales de NLTK de forma silenciosa
nltk.download('punkt', quiet=True)
nltk.download('wordnet', quiet=True)
nltk.download('omw-1.4', quiet=True)
nltk.download('stopwords', quiet=True)
nltk.download('punkt_tab')

stop_words_es = set(stopwords.words('spanish'))
stop_words = set(stopwords.words('english'))
print(stop_words_es)
```

## Paso 1: Preprocesamiento (Tokenización y Lematización)
Reducción de las palabras a su raíz morfológica básica o forma base (**lema**).

```python
texto = "The striped bats are hanging on their feet for best and are eating flies while running quickly."
texto = texto.replace("á", "a").replace("é", "e").replace("í", "i").replace("ó", "o").replace("ú", "u")
print(texto)


# Tokenizar y pasar todo a minúsculas
tokens = word_tokenize(texto.lower())

print("Texto Original:", texto)
print("Tokens:", tokens)

# limpieza
regex_expresion = r'[^a-zA-Z\s]' #--> Caracteres alfanumericos y espacios
tokens_limpios_1 = [re.sub(regex_expresion, '', token) for token in tokens]


print("Tokens Limpios:", tokens_limpios_1)

tokens_limpios = [token for token in tokens_limpios_1 if token and token not in stop_words]


print("Tokens Limpios:", tokens_limpios)


lemmatizer = WordNetLemmatizer()
Stemming = PorterStemmer()

tokens_lemmatizados = [lemmatizer.lemmatize(token) for token in tokens_limpios]

tokens_stemming = [Stemming.stem(token) for token in tokens_limpios]

print("Tokens Lemmatizados:", tokens_lemmatizados)
print("Tokens Stemming:", tokens_stemming)
```

## Paso 2: Limpieza de Texto con RegEx
Uso de expresiones regulares para eliminar caracteres especiales, números y ruido visual.

```python
reseña_sucia = "Me gustó este programa nivel 10!!! :)"

# Mantener únicamente letras del alfabeto (a-z) y espacios
reseña_limpia = re.sub(r'[^a-zA-Z\s]', '', reseña_sucia)

# Normalizar y remover espacios duplicados en blanco
reseña_limpia = " ".join(reseña_limpia.split())

print("Reseña Original:", reseña_sucia)
print("Reseña Limpia:", reseña_limpia)
```

## Paso 3: Vectorización — Bolsa de Palabras (Bag of Words)
Conversión del texto libre a vectores numéricos basados puramente en la frecuencia de ocurrencia.

```python
corpus = [
    'for want of a nail the shoe be lose',
    'for want of a shoe the horse be lose',
    'for want of a horse the rider be lose'
]

# Crear el vectorizador de frecuencias
count_vect = CountVectorizer(ngram_range=(2, 2), stop_words='english')  # Bigramas
bow_matrix = count_vect.fit_transform(corpus)

print("Vocabulario del Corpus:\n", count_vect.get_feature_names_out())
print("\nMatriz de Frecuencias (Vectores resultantes):\n", bow_matrix.toarray())
```

## Paso 4: Ponderación de Importancia con TF-IDF
Cálculo estadístico que balancea el peso de una palabra según sus apariciones locales frente a la rareza global en el dataset.

```python
corpus = [
    'for want of a nail the shoe be lose',
    'for want of a shoe the horse be lose nail',
    'for want of a horse the rider rider rider rider rider rider rider rider rider be lose nail'
]

# Aplicar TF-IDF filtrando palabras vacías habituales
tfidf_vect = TfidfVectorizer(stop_words='english')
tfidf_matrix = tfidf_vect.fit_transform(corpus)

print("Vocabulario del Corpus (TF-IDF):\n", tfidf_vect.get_feature_names_out())
print("Dimensiones de la matriz (Documentos, Vocabulario):", tfidf_matrix.shape)
print("\nVector de pesos del primer documento:\n", tfidf_matrix.toarray())
```

```python
import pandas as pd
## ejemplo final

textos_sentimientos = [
    "Me encanta este producto   ",
    "Odio este producto",
    "   !!Es un producto regular",
    "Me gusta este producto",
    "No me gusta este producto, es terrible :() ",
    "Este producto es increíble, LOOOO recomiendo mucho",
    "No recomendaría este producto a nadie, es una pérdida de dinero",
    "Este producto es DEcente <3",
    " Odio este producto   ",
    "   Me encanta este producto, es el mejor que he comprado"
]

clasificacion_textos = [1, 0, 0, 1, 0, 1, 0, 1, 0, 1] # 1: positivo, 0: negativo


df = pd.DataFrame({'Texto': textos_sentimientos, 'Sentimiento': clasificacion_textos})

df.head()


X = df['Texto']
y = df['Sentimiento']


# Limpiar
X = X.str.strip().str.lower().str.replace("á", "a").str.replace("é", "e").str.replace("í", "i").str.replace("ó", "o").str.replace("ú", "u")
X = X.apply(lambda x: re.sub(r'[^a-zA-Z\s]', '', x))
# Eliminar stopwords
X = X.apply(lambda x: " ".join([word for word in x.split() if word not in stop_words_es]))

# Tokenizar
X = X.str.split()

# lemmatizar o stemming  
lemmatizer = WordNetLemmatizer()
X = X.apply(lambda tokens: [lemmatizer.lemmatize(token) for token in tokens])

print(X)


# Vectorizar
tfidf_vect = TfidfVectorizer()
X_vectorizado = tfidf_vect.fit_transform(X.apply(lambda tokens: " ".join(tokens)))


from sklearn.linear_model import LogisticRegression

modelo = LogisticRegression()
modelo.fit(X_vectorizado, y)

prueba = ["Es un muuy mal producto, lo odio"]

prueba_vectorizada = tfidf_vect.transform(prueba)
prediccion = modelo.predict(prueba_vectorizada)
print("Predicción de Sentimiento (1=Positivo, 0=Negativo):", prediccion[0])
```
